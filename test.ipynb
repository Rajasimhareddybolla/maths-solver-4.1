```python
from openai import OpenAI
api_key = open(".openai_api_key").read().strip()
client = OpenAI(api_key=api_key)

instructions = open("system_instructions.txt").read().strip()

resp = client.responses.create(
    model="gpt-4.1",
    tools=[
        {
            "type": "code_interpreter",
            "container": {"type": "auto"}
        }
    ],
    instructions=instructions,
    input="I need to solve the equation 3x + 11 = 14. Can you help me?",
)

print(resp.output)
from openai.types.responses import ResponseCodeInterpreterToolCall, ResponseOutputMessage, ResponseOutputText

for item in resp.output:
    if isinstance(item, ResponseCodeInterpreterToolCall):
        print("--- Code Executed by the Assistant ---")
        print(item.code)
        print("--------------------------------------\n")
    elif isinstance(item, ResponseOutputMessage):
        for content_item in item.content:
            if isinstance(content_item, ResponseOutputText):
                print("--- Assistant's Response ---")
                print(content_item.text)
                print("----------------------------\n")
```
